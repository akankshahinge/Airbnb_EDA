**Data Cleaning**

```python
data.isnull().sum()
```

```python
#dropping null values
data.dropna(inplace=True)
data.isnull().sum()
```

```python
#find count of duplicate values
data.duplicated().sum()

#Print the duplicate values
data[data.duplicated()]

#Deleting duplicate values
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
#change the dataype of columns

data.dtypes
data['id'] = data['id'].astype('object')
data.dtypes
```

**EDA**

```python
#univariant analysis: analysis on one variable
#Identifying outliers in price
#sns.boxplot(data=data, x='price')
#now we will remove this outlier
df = data[data['price']<1500]
sns.boxplot(data=df, x='price')
```

```python
#histogram for price
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='price',bins=100, color='teal')
plt.ylabel('Frequency')
plt.xlabel('Price')
```

```python
#histogram for availability
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='availability_365', color='coral')
plt.ylabel('Frequency')
plt.xlabel('Price')
```

```python
#Bivariant analysis

#For bivariant analysis we will group the variable

df.groupby('neighbourhood_group')['price'].mean()
```

```python
#Get avgerage price per bed
df['price_per_bed'] = df['price']/df['beds']
df.groupby('neighbourhood_group')['price_per_bed'].mean()
```

```python
#price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
#scatter plot for reviews and price
plt.figure(figsize=(15,10))
plt.title('Price and review based on neighbourhood group')
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group', palette='magma')
```

```python
#pair plot
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365'],hue='room_type')
```

```python
#Geographical Distribution

plt.figure(figsize=(15,10))
plt.title('Geographical Distribution of Airbnb Listings')
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type', palette='viridis')
```

```python
#Heatmap
#corr is the relation of one column with another

cor = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month','availability_365','beds']].corr()
cor
sns.heatmap(cor, annot=True)
```
